```python
import signal
import json
from mflib.mf_timestamp import mf_timestamp
from fabrictestbed_extensions.fablib.fablib import FablibManager as fablib_manager
from ipaddress import ip_address, IPv4Address, IPv4Network
import matplotlib.pyplot as plt
import schedule
import time
import pandas as pd
import datetime

fablib = fablib_manager()
slice_name="file-project_" + fablib.get_bastion_username()

container_name="timestamp"
ts = mf_timestamp(slice_name=slice_name, container_name=container_name)
packet_test_name = "packet_test"
event_test_name= "event_test"


try:
    slice = fablib.get_slice(name=slice_name)
except Exception as e:
    print(f"Fail: {e}")
print (slice)
currentServer = slice.get_node(name="currentServer")
```

```python
#change routing table to move client4 from currentServer to hotServer
# set up static routes
def client4_to_hotServer():
    route_conf_to_del = [
        {"addr": "10.0.5.0/24", "gw": "10.0.2.1", "nodes": ["client3"]},
        {"addr": "10.0.2.0/24", "gw": "10.0.5.1", "nodes": ["currentServer"]},
        {"addr": "10.0.5.0/24", "gw": "10.0.3.1", "nodes": ["client4"]},
        {"addr": "10.0.3.0/24", "gw": "10.0.5.1", "nodes": ["currentServer"]},
    ]
    route_conf_to_add = [
        {"addr": "10.0.6.0/24", "gw": "10.0.2.1", "nodes": ["client3"]},
        {"addr": "10.0.2.0/24", "gw": "10.0.6.1", "nodes": ["hotServer"]},
        {"addr": "10.0.6.0/24", "gw": "10.0.3.1", "nodes": ["client4"]},
        {"addr": "10.0.3.0/24", "gw": "10.0.6.1", "nodes": ["hotServer"]},
    ]
    for rt in route_conf_to_del:
        for n in rt['nodes']:
            tmp_node = slice.get_node(name=n)
            tmp_node.ip_route_del(subnet=IPv4Network(rt['addr']), gateway=rt['gw'])
    for rt in route_conf_to_add:
        for n in rt['nodes']:
            tmp_node = slice.get_node(name=n)
            tmp_node.ip_route_add(subnet=IPv4Network(rt['addr']), gateway=rt['gw'])
    slice.get_node("client4").execute("sed -i '1s/currentServer/hotServer/' serverDST.txt")
```
