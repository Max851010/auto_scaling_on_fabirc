```python
import json
import pandas as pd
import matplotlib.pyplot as plt


file_download_path1="/home/fabric/work/final_project/current_server_data/currentServer_2024-04-20_190143.json"
file_download_path2="/home/fabric/work/final_project/current_server_data/currentServer_2024-04-20_190325.json"
file_download_path3="/home/fabric/work/final_project/current_server_data/currentServer_2024-04-20_190549.json"
file_download_path4="/home/fabric/work/final_project/current_server_data/currentServer_2024-04-20_190816.json"

with open(file_download_path1, 'r') as f:
    result1 = json.load(f)
df1 = pd.DataFrame.from_dict(result1)

with open(file_download_path2, 'r') as f:
    result2 = json.load(f)
df2 = pd.DataFrame.from_dict(result2)

with open(file_download_path3, 'r') as f:
    result3 = json.load(f)
df3 = pd.DataFrame.from_dict(result3)

with open(file_download_path4, 'r') as f:
    result4 = json.load(f)
df4 = pd.DataFrame.from_dict(result4)
```

```python

frames = [df1, df2, df3, df4]
combined_df = pd.concat(frames)
combined_df['timestamp'] = pd.to_datetime(combined_df['timestamp'])

sorted_df = combined_df.sort_values(by='timestamp')
sorted_df["timestamp"] = pd.to_datetime(sorted_df.timestamp).dt.round('1s')
plt.figure(figsize=(10, 6))
count_by_timestamp = sorted_df.groupby('timestamp').size()

count_by_timestamp.plot(kind='line', marker='o')
plt.title('Count of Packets by Timestamp')
plt.xlabel('Timestamp')
plt.ylabel('Packets')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.ylim(0, 700)
plt.show()
```
